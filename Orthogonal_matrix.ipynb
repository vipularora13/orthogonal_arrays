```python
# t=2
# l=2

comb_value=comb(n,t)

def combinations(iterable, r):
    # combinations('ABCD', 2) --> AB AC AD BC BD CD
    # combinations(range(4), 3) --> 012 013 023 123
    pool = tuple(iterable)
    n = len(pool)
    if r > n:
        return
    indices = list(range(r))
    yield tuple(pool[i] for i in indices)
    while True:
        for i in reversed(range(r)):
            if indices[i] != i + n - r:
                break
        else:
            return
        indices[i] += 1
        for j in range(i+1, r):
            indices[j] = indices[j-1] + 1
        yield tuple(pool[i] for i in indices)
i=5
nCt=combinations(mask[i],t)
#nCt = list(nCt)
nCt = np.array(list(nCt))
print(nCt)

a=nCt
combination_index=a.dot(1 << np.arange(a.shape[-1] - 1, -1, -1))
print(combination_index)
```
